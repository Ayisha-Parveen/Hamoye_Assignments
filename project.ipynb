```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("/content/understat.com.csv")
pd.set_option('display.max_columns', None)
```

```python
df.rename(columns = {"Unnamed: 0": "League","Unnamed: 1": "Year"},inplace = True)
df.head()
```

```python
print(df["League"].unique())
print(df["Year"].unique())
```

```python
team = df.groupby("team")["pts"].sum()
team
print(df["team"].nunique())
itemvalue = {}
for item in df["team"]:
  value = team[item]
  itemvalue[item] = value
print(itemvalue)

df2 = df.replace({"team":itemvalue})
df2["team"].nunique()
League = {"La_liga": 1,"EPL":2,"Bundesliga":3,"Serie_A":4,"Ligue_1":5,"RFPL":6}
df2 = df2.replace({"League":League})

position = {}
for item1 in df["position"]:
    if item1 == 1:
      value1 = 1
    else:
      value1=0
    position[item1]=value1
print(position)
df2 = df2.replace({"position":position})
df2
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df2)
df3 = scaler.transform(df2)
```

```python
df4 = pd.DataFrame(df3,columns = df.columns)
```

```python
plt.figure(figsize = [15,8])
sns.heatmap(df4.corr(),cmap = "Blues",annot = True,fmt = ".2f")
```

```python
df5 = df4.drop(["Year","League","matches","draws","missed","xG_diff","xGA_diff","ppda_coef","deep_allowed","xpts_diff"],axis = 1)
```

```python
sns.pairplot(df5,hue="position")
```

```python
inputcolumns = df5.drop("position",axis = 1)
x = inputcolumns
y = df2["position"]
```

```python
#Logistic Regression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 101)
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
predictions = logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
classifier_rf = RandomForestClassifier(random_state=42,n_estimators=10, criterion="entropy")
classifier_rf.fit(x_train, y_train)
predictions = classifier_rf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 2)
knn.fit(x_train,y_train)
pred = knn.predict(x_test)
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```
